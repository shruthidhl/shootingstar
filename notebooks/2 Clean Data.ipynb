```python
def dict_to_date(Series):
    '''Takes in a Pandas Series of dicts with the format - {'day':X, 'month':Y, 'year':Z}
    and converts them into a Pandas Series of Python date objects.'''
    dates = []
    for dicts in Series:
        if type(dicts) != dict:
            #print('nan')
            dates.append(None)
        
        else:
            i = 0
            while i < 3:
                my_lst = []
                
                for v in dicts.values():
                    my_lst.append(v)
                    i+=1
            date = dt.date(year = my_lst[-1],month = my_lst[1], day = my_lst[0])
            dates.append(date)
    
    return(pd.Series(dates))
```

```python
experience_df_final.starts_at = dict_to_date(experience_df_final.starts_at)
experience_df_final.ends_at = dict_to_date(experience_df_final.ends_at)
```
